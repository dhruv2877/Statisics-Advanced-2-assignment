1. What is hypothesis testing in statistics?
 *  Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on a sample of data. It involves formulating an assumption (called a hypothesis) about a population parameter, then using sample data to test whether this hypothesis is correct or not.

 This is done by comparing the observed data to what we would expect to see if the original assumption were correct. Based on the comparison, we either "reject" or "fail to reject" the assumption.


Example:
Let’s say a company claims that their energy drink increases reaction speed.

You want to test this, so you form hypotheses:

* H₀: The energy drink has no effect on reaction speed.

* H₁: The energy drink improves reaction speed.

You collect reaction speed data from a sample of 30 people before and after drinking the energy drink. You perform a hypothesis test using this sample data.

* If the p-value < 0.05, you reject H₀ and conclude the drink likely has an effect.

* If the p-value ≥ 0.05, you do not reject H₀ and conclude there isn’t enough evidence to prove the drink works.




2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

 *  In hypothesis testing, we work with two complementary assumptions about a population: the null hypothesis and the alternative hypothesis. These two hypotheses represent opposing views and guide how we interpret data from a sample.

* Null Hypothesis (H₀): The null hypothesis is a formal statement that there is no significant difference, effect, or relationship in the population being studied. It represents the baseline or default assumption, and statistical testing is conducted to see if there is enough evidence to reject it. Until proven otherwise, the null hypothesis is considered true.

* Alternative Hypothesis (H₁ or Ha): The alternative hypothesis proposes that there is a significant difference, effect, or relationship present in the population. It directly contradicts the null hypothesis and is accepted if the statistical evidence strongly suggests that the null hypothesis is unlikely to be true.

Differences between null hypothesis and alternative hypothesis:

* Assumption: The null hypothesis assumes no change or effect, whereas the alternative hypothesis assumes the presence of a change or effect.

* Purpose: The null hypothesis is tested to determine whether it can be rejected; the alternative hypothesis is what researchers aim to support if sufficient evidence is found.

* Outcome Based on Test Result: If the p-value is less than the significance level, the null hypothesis is rejected in favor of the alternative; otherwise, the null is retained.

3. What is the significance level in hypothesis testing, and why is it important?

 *  The significance level, denoted by α (alpha), is the probability threshold used in hypothesis testing to determine whether to reject the null hypothesis. It represents the maximum allowable probability of committing a Type I error (incorrectly rejecting a true null hypothesis). Commonly used values for α are 0.05 (5%), 0.01 (1%), and 0.10 (10%).

Importance of hypothesis testing:

* Decision Making: The significance level sets the standard for how much evidence is required to reject the null hypothesis.

* Controlling Errors: It helps control the risk of Type I errors (false positives). A lower α reduces the chance of incorrectly rejecting the null hypothesis but increases the risk of Type II errors (false negatives).

* Standardization: Provides a consistent benchmark for comparing statistical results across different studies.




4. What does a P-value represent in hypothesis testing?
 *  The P-value is a fundamental concept in hypothesis testing that quantifies the probability of observing the collected data—or something more extreme—assuming the null hypothesis (H₀) is true. It serves as a measure of the consistency between the observed data and H₀, but it does not indicate the probability that H₀ is true or false.



5. How do you interpret the P-value in hypothesis testing?

 *  The P-value is interpreted through a structured decision-making process:

1) Pre-set your significance level (α) before analyzing data. Common choices are 0.05, 0.01, or 0.10.

2) Compare P-value to α:
* If P ≤ α → Reject the null hypothesis (H₀)

* If P > α → Fail to reject H₀

3) Understand the strength of evidence:

* P < 0.01 → Very strong evidence against H₀

* 0.01 ≤ P ≤ 0.05 → Moderate evidence

* P > 0.05 → Insufficient evidence


Practical Interpretation Guide:

For a P-value of 0.03 (α = 0.05):

1) "There's a 3% probability of seeing this result if H₀ were true"

2) "This is below our 5% threshold"

3) Conclusion: "We reject H₀ in favor of the alternative"



6. What are Type I and Type II errors in hypothesis testing?

 *  In hypothesis testing, we make binary decisions that inherently carry the risk of two distinct error types. These errors represent the possible incorrect conclusions we might draw about our null hypothesis (H₀).

Type I Error (False Positive):

*  Rejecting a true null hypothesis

* Probability: Controlled by significance level (α)

* Consequences: Concluding an effect exists when it doesn't

* Control Mechanism: Set α low (typically 0.05) to limit false positives



 Type II Error (False Negative)

* Failing to reject a false null hypothesis

* Probability: Denoted by β

* Consequences: Missing a real effect

* Control Mechanism: Increase power (1-β) via larger samples or stronger effects



7.  What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
 *  Hypothesis tests can be structured to detect effects in one specific direction (one-tailed) or effects in either direction (two-tailed). This choice fundamentally changes how we allocate our "error budget" and what conclusions we can draw.

One-Tailed Test:  A directional test that evaluates the alternative hypothesis in only one specified direction.

* When to Use:

1) When only one direction of effect is meaningful

2) When theory strongly predicts a specific direction

3) When opposite-direction results would be interpreted the same as no effect

* Critical Region: Entire α is placed in one tail

* Power Advantage: More sensitive to detect effects in the specified direction

Two-tailed test: A non-directional test that evaluates deviations from H₀ in both directions.

* When to Use:

1) When effects in either direction are meaningful

2) When no strong prior expectation exists

3) Standard default approach in most research

* Critical Region: α is split between both tails (typically α/2 in each)

* Conservative Nature: Requires stronger evidence to reject H₀




8. What is the Z-test, and when is it used in hypothesis testing?

 * The Z-test is a parametric statistical hypothesis test that determines whether there is a significant difference between sample data and a population parameter (or between two sample means) by converting the data to a standard normal distribution (Z-distribution). It calculates how many standard deviations an observed value lies from the expected population mean.

 The test statistic is calculated as:

Z = (X̄ - μ) / (σ/√n)

Where:

X̄ = Sample mean

μ = Population mean

σ = Population standard deviation

n = Sample size

When to Use the Z-Test:

* Known Population Parameters: Population standard deviation (σ) is known and population mean (μ) is specified by H₀

* Sample Size Conditions: Large samples (n≥30)  or Small samples if the population is normally distributed and σ is known.

* Common Applications: Testing if a sample mean differs from a population mean, comparing two sample means (two-sample Z-test) and testing proportions (Z-test for proportions)


9.  How do you calculate the Z-score, and what does it represent in hypothesis testing?

 * The Z-score (or standard score) quantifies how many standard deviations a data point (or sample mean) is from the population mean. It standardizes values for comparison across different distributions.

Formula to calculate Z-score:

Z = (X̄ - μ) / (σ/√n)

Where:

X̄ = Sample mean

μ = Population mean

σ = Population standard deviation

n = Sample size

Role in Hypothesis Testing:

* Standardization:Converts different distributions to a common scale (standard normal distribution).

* Significance Testing: Compare calculated Z to critical values (e.g., ±1.96 for α=0.05) to decide whether to reject H₀.

* P-value Conversion: Z-scores translate directly to probabilities via the standard normal table.

 1) The sample size is small (typically n < 30),

 2) The population standard deviation (σ) is unknown,

 3) The data are approximately normally distributed.

11. What is the difference between a Z-test and a T-test?
 *  Both Z-test and T-test are statistical methods used to compare sample data with population data or between two samples. While they serve similar purposes, the choice between them depends mainly on the sample size and whether the population standard deviation is known.

* Z-test: A Z-test is a statistical test used when the population standard deviation is known, and the sample size is relatively large (typically n ≥ 30). It assumes that the sample data is approximately normally distributed. The Z-test helps determine whether the difference between the sample mean and the population mean is statistically significant.

* A T-test is used when the population standard deviation is unknown and the sample size is small (typically n < 30). It relies on the t-distribution, which accounts for the additional uncertainty in smaller samples. The T-test is commonly used to compare means either within a single sample or between two groups.

Difference between Z- test and T-test:

* Standard Deviation: The Z-test is used when the population standard deviation is known, whereas the T-test is used when it is unknown.

* Sample Size: Z-test is more suitable for large samples (n ≥ 30), while T-test is preferred for small samples (n < 30).

* Distribution Used: Z-test is based on the standard normal distribution (Z-distribution), whereas T-test uses the t-distribution which is wider and has heavier tails.

* Accuracy: The T-test adjusts for sample size variability, making it more accurate for small samples compared to the Z-test.

* Application: Z-tests are often used in quality control or large-scale surveys, while T-tests are common in experimental studies and small-sample research.



12. What is the T-test, and how is it used in hypothesis testing?
 *  A T-test is used when the population standard deviation is unknown and the sample size is small (typically n < 30). It relies on the t-distribution, which accounts for the additional uncertainty in smaller samples. The T-test is commonly used to compare means either within a single sample or between two groups.

 In hypothesis testing, the T-test helps us decide whether to reject or retain the null hypothesis based on how different the observed sample mean is from the expected mean under H₀.

13. What is the relationship between Z-test and T-test in hypothesis testing?

 *  The Z-test and T-test are both statistical methods used to test hypotheses about population means. They follow the same overall process: set up hypotheses, calculate a test statistic, and compare it to a critical value or p-value to make a decision. The main difference lies in when each test is used.

The Z-test is used when the sample size is large and the population standard deviation is known, while the T-test is used when the sample size is small and the population standard deviation is unknown.

Despite these differences, both tests serve the same purpose — to determine whether observed differences in data are statistically significant. As the sample size increases, the results of the T-test become very similar to the Z-test, since the t-distribution approaches the normal distribution.

14. What is a confidence interval, and how is it used to interpret statistical results?

 *  A confidence interval (CI) is a range of values, calculated from sample data, that is used to estimate an unknown population parameter (such as the mean or proportion). It gives an interval within which the true population value is likely to lie, along with a certain level of confidence (such as 95% or 99%).

Rather than giving a single estimate (called a point estimate), a confidence interval provides a more reliable estimate by acknowledging that there’s always some uncertainty when working with a sample.

* It Is Used to Interpret Statistical Results: confidence intervals are used to express the reliability and precision of an estimate. A wider interval suggests more uncertainty, while a narrower interval suggests greater precision.

15. What is the margin of error, and how does it affect the confidence interval?
 *  The margin of error (MoE) is a measure of the amount of uncertainty in a sample estimate. It indicates how far the sample statistic (such as the sample mean or proportion) is likely to be from the true population value.

In simple terms, it tells us how much we can expect our estimate to vary due to random sampling. It plays a critical role in determining the width of the confidence interval.

* Affects of MoE on confidence interval:

1) The confidence interval is built around the sample estimate using the margin of error:

Confidence Interval = Sample Estimate ± Margin of Error

2) A larger margin of error results in a wider confidence interval, which indicates less precision but more caution in estimation.

3) A smaller margin of error leads to a narrower confidence interval, implying greater precision but with more reliance on the sample.





16. How is Bayes’ Theorem used in statistics, and what is its significance?

 * Bayes’ Theorem is a fundamental concept in probability and statistics that describes how to update the probability of a hypothesis when new evidence or information becomes available. It provides a mathematical way to revise existing beliefs in light of new data.

Formula:

P(A|B) = (P(B|A) * P(A)) / P(B)

Where:

P(A|B): Posterior probability (probability of A given B)

P(B|A): Likelihood (probability of B given A)

P(A): Prior probability of A

P(B): Total probability of B


* Bayes’ Theorem is widely used in Bayesian statistics, where probability is interpreted as a degree of belief rather than a fixed frequency. It is especially important in situations where new information continuously updates our understanding.

Significance of baye's theorem in statistics:

1) It allows incorporation of prior knowledge with observed data

2) It gives a dynamic approach to probability instead of a static one

3) It is foundational in many modern statistical models and AI systems





17. What is the Chi-square distribution, and when is it used?

 *  The Chi-square distribution (χ² distribution) is a statistical distribution that arises when the sum of the squares of independent standard normal variables is calculated. It is a continuous probability distribution and is used primarily in tests of categorical data.

The Chi-square distribution is mainly used in the following scenarios:

* Goodness-of-Fit Test:
To determine whether observed data follows a specific theoretical distribution (e.g., checking if dice are fair).

* Test of Independence:
To examine the association between two categorical variables in a contingency table (e.g., relationship between gender and voting preference).

* Test of Homogeneity:
To compare distributions of a categorical variable across different populations.

* Testing Variance:
In some cases, it is also used to test the variance of a normally distributed population.

18. What is the Chi-square goodness of fit test, and how is it applied?

 *  The Chi-square goodness of fit test is a statistical method used to determine if a set of observed frequencies (or counts) matches the expected frequencies based on a specific hypothesis. It is particularly useful for categorical data to check if the distribution of data fits a certain expected pattern or model.

The test involves comparing the observed data with the expected data under the assumption of the null hypothesis (which typically states that there is no significant difference between the observed and expected frequencies).

The Chi-square statistic is calculated using the formula:

χ² = Σ ((O - E)² / E)

Where:
O = Observed frequency
E = Expected frequency


19.  What is the F-distribution, and when is it used in hypothesis testing?

 *  The F-distribution is a probability distribution that arises in the context of comparing variances and is used primarily in analysis of variance (ANOVA), regression analysis, and hypothesis testing for the ratio of two variances. It is a continuous probability distribution that is positively skewed, starting at 0 and extending indefinitely to the right. The shape of the F-distribution depends on two parameters: the degrees of freedom (df) of the numerator and the degrees of freedom of the denominator.

 Use of F-distribution in hypothesis testing:

* Analysis of Variance (ANOVA):
To test if there are significant differences between the means of multiple groups.

* Regression Analysis:
To test the overall significance of a regression model.

* Comparing Two Variances:
To test if two population variances are equal.

*  Homogeneity of Variance:
To check if multiple samples have the same variance.

20. What is an ANOVA test, and what are its assumptions?
 *  ANOVA (Analysis of Variance) is a statistical test used to compare the means of three or more groups to determine if there are any statistically significant differences among them. It works by analyzing the variance within each group and comparing it to the variance between the groups. If the variance between the groups is significantly larger than the variance within the groups, it suggests that at least one of the group means is different from the others.

 Assumptions of ANOVA test:

* Independence of Observations:
Data points in each group must be independent of each other.

* Normality:
Data in each group should be approximately normally distributed.

* Homogeneity of Variance:
The variances of the groups should be roughly equal.

* Random Sampling:
The samples should be randomly selected from the population.

21.  What are the different types of ANOVA tests?
 *  ANOVA (Analysis of Variance) has different types based on the number of factors (independent variables) being analyzed and the structure of the data. These types help in analyzing the variance in different experimental designs. The main types of ANOVA are:

* One-Way ANOVA:
Used when there is one factor (independent variable) with three or more levels (groups) to compare.

* Two-Way ANOVA:
Used when there are two factors to analyze. It checks for the effect of each factor individually and also whether there is an interaction between the two factors.

* Repeated Measures ANOVA:
Used when the same subjects are used for each treatment (i.e., repeated measurements). It accounts for the correlation between observations of the same subject.

* Multivariate Analysis of Variance (MANOVA):
An extension of ANOVA that allows for the analysis of multiple dependent variables simultaneously to determine if they are influenced by the independent variables.

* Factorial ANOVA:
A generalization of two-way ANOVA that deals with multiple factors, allowing for testing the effect of each factor, their interactions, and combined effects on the dependent variable.



22. What is the F-test, and how does it relate to hypothesis testing?

 *  The F-test is a statistical test used to compare two variances to determine if they are significantly different. It is based on the F-distribution and calculates the ratio of two sample variances. The test helps assess whether the variability between groups is larger than the variability within groups, which is important in analysis of variance (ANOVA) and other statistical methods.

 In hypothesis testing, the F-test helps determine whether the observed variances are due to random chance or if there is a significant difference between the groups' variability. It is commonly used in:

* ANOVA: To test the differences in means across multiple groups.
* Regression: To test if the model explains a significant portion of the variability in the data.

```python
# 2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

import numpy as np
import matplotlib.pyplot as plt

low = 1      # Minimum value
high = 6     # Maximum value (inclusive)
size = 1000  # Number of samples

# Generating discrete uniform distribution (like rolling a die)
data = np.random.randint(low, high + 1, size=size)

# Calculating PMF (empirical probabilities)
unique, counts = np.unique(data, return_counts=True)
pmf = counts / size

# Plotting PMF
plt.bar(unique, pmf, color='skyblue', edgecolor='black')
plt.xlabel('Outcomes (e.g., die faces)')
plt.ylabel('Probability')
plt.title('PMF of Discrete Uniform Distribution (1-6)')
plt.xticks(unique)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

import numpy as np
import matplotlib.pyplot as plt

def bernoulli_pdf(p):
    """
    Calculate and plot the PDF of a Bernoulli distribution.

    Parameters:
        p (float): Probability of success (1), where 0 <= p <= 1.
    """
    # Possible outcomes (0 = failure, 1 = success)
    outcomes = np.array([0, 1])
    probabilities = np.array([1 - p, p])

    # Plotting PDF
    plt.bar(outcomes, probabilities, color=['red', 'green'], alpha=0.7)
    plt.xticks(outcomes, labels=['Failure (0)', 'Success (1)'])
    plt.ylabel('Probability')
    plt.title(f'Bernoulli Distribution PDF (p={p})')
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.ylim(0, 1)
    plt.show()

p = 0.3  # Probability of success (e.g., 30% chance)
bernoulli_pdf(p)
```

```python
# 4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 10  # Number of trials
p = 0.5 # Probability of success
size = 1000 # Number of simulations

data = np.random.binomial(n, p, size)

# Plotting histogram
plt.hist(data, bins=range(n+2), align='left', rwidth=0.8, edgecolor='black', density=True)

x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)
plt.plot(x, pmf, 'ro-', markersize=5, label='Theoretical PMF')

# Formatting
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.legend()
plt.grid(axis='y', alpha=0.5)
plt.xticks(x)
plt.show()
```

```python
#5. Create a Poisson distribution and visualize it using Python.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

lam = 3  # Average rate of occurrence (lambda)
size = 1000  # Number of samples

# Generating Poisson distribution
data = np.random.poisson(lam, size)

# Creating histogram
plt.hist(data, bins=np.arange(-0.5, 15.5, 1),
         density=True, alpha=0.7,
         edgecolor='black', label='Simulated Data')

x = np.arange(0, 15)
pmf = poisson.pmf(x, lam)
plt.plot(x, pmf, 'ro-', markersize=5, label='Theoretical PMF')

# Formatting
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.title(f'Poisson Distribution (λ={lam})')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.xticks(x)
plt.show()
```

```python
# 6. Write a Python program to calculate and plot the cumulative distribution
# function (CDF) of a discrete uniform distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

low = 1
high = 6
n_samples = 1000

# Generating discrete uniform data (like rolling a die)
data = np.random.randint(low, high + 1, size=n_samples)

# Calculating empirical CDF
sorted_data = np.sort(data)
y_vals = np.arange(1, len(sorted_data) + 1) / len(sorted_data)

# Calculating theoretical CDF
x_theoretical = np.arange(low, high + 2)  # +2 to extend the last step
cdf_theoretical = randint.cdf(x_theoretical, low, high + 1)

# Plotting empirical vs theoretical CDF
plt.step(sorted_data, y_vals, where='post', label='Empirical CDF')
plt.step(x_theoretical, cdf_theoretical, where='post',
         linestyle='--', color='red', label='Theoretical CDF')

# Formatting
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.title(f'CDF of Discrete Uniform Distribution ({low}-{high})')
plt.legend()
plt.grid(alpha=0.3)
plt.xticks(np.arange(low, high + 1))
plt.ylim(0, 1.1)
plt.show()
```

```python
# 7. Generate a continuous uniform distribution using NumPy and visualize it.

import numpy as np
import matplotlib.pyplot as plt

low = 2.0    # Lower bound
high = 5.0   # Upper bound
size = 1000  # Number of samples

# Generating continuous uniform distribution
data = np.random.uniform(low, high, size)

# Plotting histogram
plt.hist(data, bins=30, density=True, alpha=0.7, edgecolor='black')

# Plotting theoretical PDF
x = np.linspace(low-1, high+1, 1000)
pdf = np.where((x >= low) & (x <= high), 1/(high-low), 0)
plt.plot(x, pdf, 'r-', linewidth=2, label='Theoretical PDF')

# Formatting
plt.xlabel('Value')
plt.ylabel('Density')
plt.title(f'Continuous Uniform Distribution ({low} to {high})')
plt.legend()
plt.grid(alpha=0.3)
plt.xlim(low-1, high+1)
plt.show()
```

```python
#8. Simulate data from a normal distribution and plot its histogram.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 0      # Mean
sigma = 1    # Standard deviation
size = 1000  # Number of samples

# Generating normal distribution data
data = np.random.normal(mu, sigma, size)

# Plotting histogram
plt.hist(data, bins=30, density=True, alpha=0.7,
         edgecolor='black', label='Sampled Data')

# Plotting theoretical PDF
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, 'r-', linewidth=2, label='Theoretical PDF')

# Formatting
plt.xlabel('Value')
plt.ylabel('Density')
plt.title(f'Normal Distribution (μ={mu}, σ={sigma})')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a given dataset.

    Parameters:
        data (array-like): Input data array

    Returns:
        z_scores (numpy array): Standardized Z-scores
    """
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z_scores = (data - mean) / std
    return z_scores

np.random.seed(42)
original_data = np.random.normal(loc=50, scale=10, size=100)

# Calculate Z-scores
z_data = calculate_z_scores(original_data)

plt.figure(figsize=(12, 5))

# Plotting original data
plt.subplot(1, 2, 1)
plt.hist(original_data, bins=15, color='skyblue', edgecolor='black')
plt.title('Original Data')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(alpha=0.3)

# Plotting standardized data
plt.subplot(1, 2, 2)
plt.hist(z_data, bins=15, color='lightgreen', edgecolor='black')
plt.title('Standardized Data (Z-scores)')
plt.xlabel('Z-scores')
plt.ylabel('Frequency')
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Print some statistics
print(f"Original Data - Mean: {np.mean(original_data):.2f}, Std: {np.std(original_data, ddof=1):.2f}")
print(f"Z-scores - Mean: {np.mean(z_data):.2f}, Std: {np.std(z_data, ddof=1):.2f}")
```

```python
#5.  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing


import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def type_errors(population_mean, sample_size, population_std, alpha=0.05):
    np.random.seed(0)
    sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
    sample_mean = np.mean(sample_data)

    # Z-statistic for the null hypothesis
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    z_critical = stats.norm.ppf(1 - alpha / 2)  # Critical value for two-tailed test
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))  # Two-tailed P-value

    # Type 1 Error: Rejecting H0 when H0 is true
    type_1_error = abs(z_statistic) > z_critical

    # Simulate new sample data for the alternative hypothesis
    new_population_mean = population_mean + 5  # Adjusted population mean
    new_sample_data = np.random.normal(loc=new_population_mean, scale=population_std, size=sample_size)
    new_sample_mean = np.mean(new_sample_data)

    # Z-statistic under the alternative hypothesis
    new_z_statistic = (new_sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Type 2 Error: Failing to reject H0 when H0 is false
    type_2_error = abs(new_z_statistic) < z_critical

    # Visualize the results
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label="Z-distribution", color="blue")
    plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color="red", alpha=0.5, label="Type 1 Error Region")
    plt.fill_between(x, y, where=(x >= -z_critical) & (x <= z_critical), color="yellow", alpha=0.5, label="Type 2 Error Region")
    plt.axvline(z_statistic, color="green", linestyle="--", label=f"Z-statistic (H0): {z_statistic:.2f}")
    plt.axvline(new_z_statistic, color="purple", linestyle="--", label=f"Z-statistic (H1): {new_z_statistic:.2f}")
    plt.title("Type 1 and Type 2 Error Visualization")
    plt.xlabel("Z-values")
    plt.ylabel("Probability Density Function")
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Sample Mean: {sample_mean}")
    print(f"Z-statistic (H0): {z_statistic}")
    print(f"Z-statistic (H1): {new_z_statistic}")
    print(f"P-value: {p_value}")

    if type_1_error:
        print("Type 1 Error: Rejecting H0 when H0 is actually true.")
    else:
        print("No Type 1 Error.")

    if type_2_error:
        print("Type 2 Error: Failing to reject H0 when H0 is actually false.")
    else:
        print("No Type 2 Error.")

population_mean = 50
sample_size = 100
population_std = 10

type_errors(population_mean, sample_size, population_std)
```

```python
#8.  Simulate data and perform both Z-test and T-test, then compare the results using Python

import numpy as np
from scipy import stats

def perform_z_test(sample, population_mean, population_std, sample_size):
    # Z-test: Using population standard deviation
    sample_mean = np.mean(sample)
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))  # Two-tailed p-value
    return z_statistic, p_value

def perform_t_test(sample, population_mean, sample_size):
    # T-test: Using sample standard deviation
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimator)
    t_statistic, p_value = stats.ttest_1samp(sample, population_mean)
    return t_statistic, p_value

def compare_tests(population_mean, population_std, sample_size):
    # Simulate a sample dataset
    sample_data = np.random.normal(population_mean, population_std, sample_size)

    # Perform Z-test
    z_statistic, z_p_value = perform_z_test(sample_data, population_mean, population_std, sample_size)

    # Perform T-test
    t_statistic, t_p_value = perform_t_test(sample_data, population_mean, sample_size)

    print(f"Z-test result:")
    print(f"Z-statistic: {z_statistic}")
    print(f"P-value: {z_p_value}")

    print("\nT-test result:")
    print(f"T-statistic: {t_statistic}")
    print(f"P-value: {t_p_value}")

    if z_p_value < 0.05:
        print("\nZ-test: Reject the null hypothesis. The sample mean is significantly different from the population mean.")
    else:
        print("\nZ-test: Fail to reject the null hypothesis. The sample mean is not significantly different from the population mean.")

    if t_p_value < 0.05:
        print("T-test: Reject the null hypothesis. The sample mean is significantly different from the population mean.")
    else:
        print("T-test: Fail to reject the null hypothesis. The sample mean is not significantly different from the population mean.")

population_mean = 50
population_std = 10
sample_size = 100

compare_tests(population_mean, population_std, sample_size)
```

```python
#9.  Write a Python function to calculate the confidence interval for a sample mean and explain its significance


import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence_level=0.95):
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    sample_size = len(sample)

    # Z-value or t-value (depending on the confidence level and sample size)
    if sample_size >= 30:  # Use Z-distribution for large sample sizes (n > 30)
        critical_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)
    else:  # Use T-distribution for small sample sizes
        critical_value = stats.t.ppf(1 - (1 - confidence_level) / 2, df=sample_size - 1)

    # Margin of error
    margin_of_error = critical_value * (sample_std / np.sqrt(sample_size))

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

sample_data = np.random.normal(50, 10, 100)

lower, upper = confidence_interval(sample_data)

print(f"Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
#11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process


def bayesian_inference(prior, likelihood, evidence):
    # Bayes' Theorem: P(A | B) = (P(B | A) * P(A)) / P(B)
    posterior = (likelihood * prior) / evidence
    return posterior

# Example values:
# P(A) = Prior probability (probability of having the disease)
prior = 0.01  # 1% of the population has the disease

# P(B | A) = Likelihood (probability of a positive test given that you have the disease)
likelihood = 0.95  # 95% test accuracy

# P(B) = Marginal likelihood (probability of a positive test)
# P(B) is calculated as:
# P(B) = P(B | A) * P(A) + P(B | not A) * P(not A)
# where:
# P(B | not A) = false positive rate (probability of a positive test given you don't have the disease)
# P(not A) = 1 - P(A)

false_positive_rate = 0.05  # 5% false positive rate
not_A = 1 - prior  # Probability of not having the disease

# Calculating P(B) (Total probability of a positive test)
evidence = (likelihood * prior) + (false_positive_rate * not_A)

# Calculate posterior (the probability of having the disease given a positive test result)
posterior = bayesian_inference(prior, likelihood, evidence)

print(f"Posterior Probability (Probability of having the disease given a positive test): {posterior:.4f}")
```

```python
#16.  Implement an F-test using Python to compare the variances of two random samples


import numpy as np
from scipy.stats import f

sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=52, scale=5, size=30)

# Calculate sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# F-statistic (larger variance / smaller variance)
if var1 > var2:
    f_stat = var1 / var2
    dfn, dfd = len(sample1) - 1, len(sample2) - 1
else:
    f_stat = var2 / var1
    dfn, dfd = len(sample2) - 1, len(sample1) - 1

# Calculate the critical value for 95% confidence
alpha = 0.05
f_critical = f.ppf(1 - alpha, dfn, dfd)

print("Variance of Sample 1:", var1)
print("Variance of Sample 2:", var2)
print("F-statistic:", f_stat)
print("F-critical value (95% confidence):", f_critical)

if f_stat > f_critical:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
```

```python
#19.  Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

import numpy as np
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups):
    # Check normality using Shapiro-Wilk test
    for i, group in enumerate(groups, 1):
        stat, p = shapiro(group)
        print(f"Group {i} - Shapiro-Wilk Test p-value: {p}")
        if p > 0.05:
            print("  Likely Normal")
        else:
            print("  Not Normal")

    # Check homogeneity of variances using Levene’s test
    stat, p = levene(*groups)
    print(f"\nLevene’s Test p-value: {p}")
    if p > 0.05:
        print("Equal variances assumed.")
    else:
        print("Variances are significantly different.")

    print("\nNote: Independence is assumed if data was randomly sampled.")

# Example usage
group1 = np.random.normal(60, 4, 30)
group2 = np.random.normal(62, 4, 30)
group3 = np.random.normal(67, 4, 30)

check_anova_assumptions(group1, group2, group3)
```

```python
#24.  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results


import numpy as np
from scipy.stats import chi2

np.random.seed(0)
sample_data = np.random.normal(loc=50, scale=10, size=100)

# Population variance
population_variance = 100

# Sample variance
sample_variance = np.var(sample_data, ddof=1)

# Perform Chi-square test for population variance
n = len(sample_data)  # sample size
chi_square_stat = (n - 1) * sample_variance / population_variance

# Degrees of freedom
df = n - 1

# P-value for two-tailed test
p_value = 2 * min(chi2.cdf(chi_square_stat, df), 1 - chi2.cdf(chi_square_stat, df))

print(f"Chi-square statistic: {chi_square_stat}")
print(f"P-value: {p_value}")

# Hypothesis Testing Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The sample variance is significantly different from the population variance.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample variance and population variance.")
```
